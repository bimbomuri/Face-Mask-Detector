```
import numpy as np
import cv2
from keras.models import load_model
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```
model = load_model('maskmdel.hdf5')
model.summary()
```

```
def show_img(img_path):
    img = cv2.imread(img_path)
    cv_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(cv_rgb)
```

```
def preprocessing(img_path): 
     image = cv2.imread(img_path)
     image = cv2.resize(image,(224,224))/255.0
     dim_image = np.expand_dims(image, axis=0)
     dim_image = np.array(dim_image,dtype='float32')
     return dim_image
   
  
```

```
def prediction(dim_image):
   dim_image = preprocessing(img_path)
   predicted_vector = model.predict(dim_image)
   predicted_vector1 = np.argmax(predicted_vector)>0.5
   if predicted_vector1 == True:
        print('No FaceMask Detected!')
   if predicted_vector1 == False :
        print('Face Mask Detected!')
```

```
img_path ='/content/74967117_780116112438006_7313471599932276805_n(2).jpg'
preprocessing(img_path)
prediction(img_path)
show_img(img_path)
```
